```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://github.com/mojones/binders/raw/master/weather.csv')
df.tail()
```

```python
data = df[(df['year'] == 2000) & (df['city'] == 'edinburgh')]
sns.catplot(
    data=data,
    x='month',
    y='temp',
    aspect = 2,
    color='red',
    kind='swarm'
)
plt.title('Mean daily temperature in Edinburgh in 2000 (swarm plot)')
plt.savefig('swarm_2000.png', dpi=300)
```
